```python
import pandas as pd
import numpy as np
from numpy import nan as NA

df = pd.read_csv('bikes_borrowed.csv')
print(df.shape)
df.head()
```

    نحوه‌ی کار با تابع <code>read_csv</code> را در درسنامه‌های آینده یاد می‌گیرید؛ اما به استفاده از <code>head</code> در این کد دقت کنید؛ همان‌طور که گفتیم، این تابع به‌طور پیش‌فرض، ۵ داده‌ی بالای دیتافریم را به ما نشان می‌دهد.

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت اول
    </font>
</h2>


<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    همان‌طور که می‌بینید نام ستون‌های <code>t1</code> و <code>t2</code> خیلی کلی است و بیانگر معنای مقدار آن‌ها نیست، به همین دلیل خوب است تا نام آن‌ها را تغییر دهیم.
<br>
    در سلول زیر، نام ستون <code>t1</code> را به <code>t_real</code> و نام ستون <code>t2</code> را به <code>t_feels_like</code> تغییر دهید.
<br>
    برای تغییر نام نمایه‌ها یا ستون‌ها می‌توانیم همچون کد زیر از تابع <code>rename</code> استفاده کنیم:
<br>
</font>
</p>

`df.rename(columns={"col1": "new_col1", "col2": "new_col2"} , inplace = True)`


```python
df.rename(columns={'t1':"t_real",'t2':"t_feels_like"},inplace = True)

df.head()
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    به پارامتر <code>inplace = True</code> دقت کنید؛ وجود این پارامتر باعث می‌شود که تابع <code>rename</code>، به‌جای آن‌که یک کپی از دیتافریم تهیه کرده و بعد از اعمال تغییرات آن کپی را خروجی دهد، تغییرات را مستقیماً روی دیتافریم اصلی اعمال کند.
<br>
برای درک بهتر این موضوع می‌توانید یک‌بار این پارامتر را حذف کرده و سپس مجدد <code dir=ltr>df.head()</code> را خروجی بگیرید و بررسی کنید که آیا نام ستون‌ها تغییر کرده است یا خیر؟
</font>
</p>


<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت دوم
    </font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
از آن‌جایی که اغلب روز‌هایی که دوچرخه‌سواران از هوا گلایه دارند روز‌های بادی است، تصمیم گرفتیم که محاسبات را فقط برای روز‌هایی که سرعت باد بیشتر از ۱۰ است انجام دهیم.
<br>
    بنابراین در دیتافریم <code>windy_days_df</code>، تنها داده‌هایی را ذخیره کنید که سرعت باد آن‌ها بیشتر از ۱۰ باشد (خود ۱۰ را در نظر نگیرید).
</font>
</p>

```python
windy_days_df = df.wind_speed > 10.0
# print(windy_days_df.head)
windy_days_df.head()
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت سوم
    </font>
</h2>


<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
برای انجام باقی محاسبات فقط ستون‌های مخصوص دما را نیاز داریم؛ پس برای دیتافریم <code>windy_days_df</code>، ستون‌های <code>humidity</code> , <code>t_feels_like</code> , <code>t_real</code> و <code>wind_speed</code> را ذخیره کرده و باقی ستون‌ها را نادیده بگیرید.

</font>
</p>

```python
windy_days_df = df[["t_real","t_feels_like","humidity","wind_speed"]][df.wind_speed > 10.0]
# windy_days_df = df[df.wind_speed > 10.0]

windy_days_df.head()
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در مراحل بعدی می‌خواهیم برخی مقادیر دیتافریم را تغییر دهیم پس برای آن‌که دیتافریم اصلی و <code>windy_days_df</code> تغییر نکنند لازم است تا از تابع <code>copy</code> استفاده کنیم.
<br>
به‌طور کلی استفاده از عملگر مساوی برای دیتافریم‌ها در پانداز مانند نامپای عمل می‌کنند و تنها یک اشاره‌گر (pointer) جدید به همان دیتافریم اضافه می‌شود. بنابراین، تغییر دادن هر کدام از دیتافریم‌ها باعث تغییر دیگری نیز می‌شود.
<br>
    استفاده از تابع <code>copy</code> این قابلیت را برای ما ایجاد می‌کند که یک کپی از دیتافریم اولیه دریافت کنیم و تغییر یکی تاثیری روی دیگری نداشته باشد.
<br>
در ادامه بیشتر  این ویژگی را بررسی می‌کنیم اما فعلاً کافیست سلول زیر را اجرا کنید.
</font>
</p>

```python
temperature_df = windy_days_df.copy()
temperature_df
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت چهارم
    </font>
</h2>



<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
برای انجام تحلیل‌های بهتر نیاز داریم بدانیم که یک روز نسبت به بازده‌ی کلی دما تا چه‌اندازه گرم است. برای این کار به‌ترتیب مراحل زیر را انجام دهید:
<br>
    ۱- در متغیر <code>t_max</code> مقدار ماکزیمم ستون <code>t_real</code> را ذخیره کنید.
<br>
    ۲- در متغیر <code>t_min</code> مقدار مینیمم ستون <code>t_real</code> را ذخیره کنید.
<br>
۳- ستون جدیدی به نام <code>t_percent</code> به دیتافریم اضافه کنید. در این ستون از فرمول زیر برای نرمال‌سازی و پیدا کردن دمای نسبی استفاده کنید.
<br>
</font>
</p>

`((temp - min) / (max - min)) * 100`

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
<span style="color:green"><b>راهنمایی:</b></span>
 تمام عملیات این مرحله را می‌توانید با توابع نامپای انجام دهید. همان‌طور که گفتیم پانداز بر اساس نامپای نوشته شده، پس تمامی عملیات‌های نامپای در این‌جا هم قابل استفاده است.
</font>
</p>

```python
t_max = np.max(temperature_df["t_real"])
print(t_max)
t_min = np.min(temperature_df["t_real"])
print(t_min)
# print(temperature_df.head())
print(temperature_df["t_real"])
temperature_df['t_percent'] = ((temperature_df["t_real"] - t_min) / (t_max - t_min)) * 100

temperature_df.head()
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت پنجم
    </font>
</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
حال می‌خواهیم دمایی که احساس می‌شود را طبق اطلاعات موجود پیش‌بینی کرده و مقادیر این ستون (<code>t_feels_like</code>) را جایگزین کنیم.
برای مقداردهی این ستون از فرمول زیر استفاده کنید:
<br>
</font>
</p>

`t_feels_like = t_real + (humidity * t_real)/1000 - (wind_speed)/10 -2`
<br>
<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    <span style="color:orange"><b>یادآوری:</b></span>
 هر ستون دیتافریم را می‌توانید مانند یک سری جدا کرده و مانند یک آرایه‌ی نامپای با آن رفتار کنید.
</font>
</p>

```python
t_feels_like = temperature_df["t_real"] + (temperature_df["humidity"] * temperature_df["t_real"])/1000 - (temperature_df["wind_speed"])/10 -2
print(t_feels_like.head())
temperature_df['t_feels_like'] = t_feels_like

temperature_df.head()
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    همان‌طور که می‌بینید مقدار ستون <code>t_feels_like</code> تغییر کرده و دیگر مانند آن‌چه در دیتافریم اولیه داشتیم نیست. حال بیایید کارکرد تابع ‍<code>copy</code> را بررسی کنیم. اگر تمام مراحل را درست پیش رفته باشید، مقدار ستون <code>t_feels_like</code> در دیتافریم ‍‍<code>windy_days_df</code> نباید تغییر کرده باشد. برای اطمینان از این موضوع با استفاده از تابع <code>head</code> می‌توانیم مقادیر این دو دیتافریم را مقایسه کنیم.

</font>
</p>

```python
windy_days_df.head()
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت ششم
    </font>
</h2>



<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در نهایت می‌خواهیم دقت این پیش‌بینی را محاسبه کنیم. برای این کار از فرمول زیر استفاده می‌کنیم که با نام <i>Mean Absolute Error</i> یا <i>MAE</i> شناخته می‌شود.
<br>
</font>
</p>

`np.mean(np.abs(temperature_df['t_feels_like'] - windy_days_df['t_feels_like']))`
<br>
<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    این مقدار را محاسبه و در متغیر <code>diffrence</code> ذخیره کنید.
</font>
</p>

```python
diffrence = np.mean(np.abs(temperature_df['t_feels_like'] - windy_days_df['t_feels_like']))
print(diffrence)
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    در این تمرین با تابع <code>head(n)</code> کار کردیم اما بیایید به بررسی عمیق‌تر این تابع بپردازیم.
<br>
به‌طور کلی این تابع تنها برای نمایش داده‌ها استفاده نمی‌شود، بلکه اعمال آن روی دیتافریم یک کپی از <code>n</code> داده‌ی اول آن را برمی‌گرداند و می‌توان آن را در دیتافریم جداگانه‌ای ذخیره کرد.
<br>
    در دیتافریم زیر، ۱۰۰ داده‌ی اول دیتافریم <code>temperature_df</code> را ذخیره کنید.

</p>

```python
final_df = temperature_df.head(100)
print(final_df)
```

    برای ساخته‌شدن فایل <code>result.zip</code> سلول زیر را اجرا کنید. توجه داشته باشید که پیش از اجرای سلول زیر تغییرات اعمال شده در نت‌بوک را ذخیره کرده باشید (<code>ctrl+s</code>) تا در صورت نیاز به پشتیبانی امکان بررسی کد شما وجود داشته باشد.

```python
import zlib
import zipfile

np.savez("answers.npz", diffrence=diffrence, t_min=t_min , t_max=t_max)
df.to_csv('bikes.csv', index=True)
windy_days_df.to_csv('windy_days.csv', index=True)
temperature_df.to_csv("temperature.csv", index=True)
final_df.to_csv("final.csv", index=True)

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

file_names = ["bikes.csv", "windy_days.csv", "temperature.csv", 
              "final.csv", "answers.npz" , "mad_cyclists.ipynb"]
compress(file_names)
```
